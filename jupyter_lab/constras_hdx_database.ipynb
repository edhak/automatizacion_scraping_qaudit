```python
#dict_data['proyecto'] = 245169
def get_excelinfo_seg(df_excel_hds,ol):
    dict_data = {
        'proyecto':ol,
        'source':'excel',
        'invitaciones_seg':0,
        'provedores_ruc_seg':0,
        'proveedores_name_seg':0,
        'iniciaron_proceso_seg':0,
        'no_participa_seg':0,
        'otros_estados_seg':0,
        
        }
    #dict_data['source'] = 'excel'
    dict_data['invitaciones_seg'] = df_excel_hds.shape[0]
    dict_data['provedores_ruc_seg'] = len(df_excel_hds.groupby(df_excel_hds['NEW_RUC'].to_list()).size())
    dict_data['proveedores_name_seg'] = len(df_excel_hds.groupby(df_excel_hds['Proveedor'].to_list()).size())
    
    estados_count = df_excel_hds.groupby(df_excel_hds['NEW_ESTADO'].to_list()).size()
    
    dict_data['iniciaron_proceso_seg'] = estados_count[8] #estado 8 es 'INICIÓ EL PROCESO' 
    dict_data['no_participa_seg'] = estados_count[3] #estado 8 es 'NO VA A PARTICIPAR'
    
    estados_restantes_count = dict_data['invitaciones_seg'] - (dict_data['iniciaron_proceso_seg'] + dict_data['no_participa_seg'])
    dict_data['otros_estados_seg'] = estados_restantes_count
    return dict_data

get_excelinfo_seg(df_excel_single,'245169')
```

```python
def get_excelinfo_hdr(df_excel_hdr):
    dict_dat = {
        'iniciaron_proceso_hdr':0,
        'entregados_hdr':0,
        'otros_estados_hdr':0,
         'linea_adicional': 0,
         'lineas_principal': 0,
        'evaluaciones_elv':0,
        'avg_puntaje_elv':0,
        'informes_inf':0
    }
    
    dict_dat['iniciaron_proceso_hdr'] = df_excel_hdr.shape[0]
    count_estados = prueba_hdr.groupby(prueba_hdr['NEW_ESTADO'].to_list()).size()
    dict_dat['entregados_hdr'] = count_estados[5] # 5 es el estado de estregado
    dict_dat['otros_estados_hdr'] = dict_dat['iniciaron_proceso_hdr'] - dict_dat['entregados_hdr']
    #'lineas_principal','linea_adicional'
    dict_dat['linea_adicional'] = prueba_hdr[(prueba_hdr['NEW_ESTADO'] == 5) & (prueba_hdr['Tipo_Evaluacion'] == 'LINEA ADICIONAL')]['NEW_ESTADO'].count()
    dict_dat['lineas_principal'] = dict_dat['entregados_hdr'] - dict_dat['linea_adicional']
    dict_dat['evaluaciones_elv'] = count_estados[5] # 5 es el estado también es la cantidad de evaluaciones
    dict_dat['avg_puntaje_elv'] = int(prueba_hdr[prueba_hdr['NEW_ESTADO'] == 5]['Puntaje'].mean())  #solo cogemos el entero
    dict_dat['informes_inf'] = count_estados[5] # 5 es el estado también es la cantidad de evaluaciones
    
    return dict_dat

get_excelinfo_hdr(prueba_hdr)
```

```python
def get_excel_info(df_excel_hds, df_excel_hdr, ol):
    #get_excelinfo_seg(df_excel_hds, ol)
    #get_excelinfo_hdr(df_excel_hdr)
    return {**get_excelinfo_seg(df_excel_hds,ol),**get_excelinfo_hdr(df_excel_hdr)}

get_excel_info(df_excel_single,prueba_hdr,'245169')
```

```python
def get_data_db():
    query_seg = """
            select ocmnum, ciaidenuminv, ciades, maedetcod, maedetdes from hom.TbUltimoSeguimiento
            where ocmnum in (
                select ocmnum from hom.tbOrdenesComerciales
                where
                dotemicod = '70933'
                and ocmest <> 0
                and isnull(ocmnum,'') <> ''
                and ocmfehcre >= convert(date,dateadd(year,-3,getdate()))
            )
        """
    df_db_seg = pd.read_sql_query(query_seg,db) 
    #data_seg.head()
    
    query_hdr = """
            select ocmnum, oinnum,oinref,ciaidenum_mandante, ciaidenum_proveedor, maecabestcod,maecabtipcod  from hom.TbOrdenesInspeccion
            where ocmnum in (
                    select ocmnum from hom.tbOrdenesComerciales
                    where
                    dotemicod = '70933'
                    and ocmest <> 0
                    and isnull(ocmnum,'') <> ''
                    and ocmfehcre >= convert(date,dateadd(year,-3,getdate()))
                )
        """
    df_db_hdr = pd.read_sql_query(query_hdr,db) 
    
    query_evl = """
        select ocmnum,isnull(avg(evlpje),0) as puntaje ,count(ciacod) as num_eval from hom.TbEvaluaciones
        where ocmnum in (
                select ocmnum from hom.tbOrdenesComerciales
                where
                dotemicod = '70933'
                and ocmest <> 0
                and isnull(ocmnum,'') <> ''
                and ocmfehcre >= convert(date,dateadd(year,-3,getdate()))
        )
        group by ocmnum
        """
    df_db_evl = pd.read_sql_query(query_evl,db) 
    
    query_inf = """
        select ocmnum,count(distinct(evlcod)) as num_inf from hom.TbInformes
        where ocmnum in (
                select ocmnum from hom.tbOrdenesComerciales
                where
                dotemicod = '70933'
                and ocmest <> 0
                and isnull(ocmnum,'') <> ''
                and ocmfehcre >= convert(date,dateadd(year,-3,getdate()))
        )
        group by ocmnum
        """
    #df_db_inf = pd.read_sql_query(query_inf,db) 

    return {'df_db_seg':df_db_seg, 'df_db_hdr':df_db_hdr, 'df_db_evl':df_db_evl}

#get_data_db()
```

```python
def get_db_info(df_db_seg,df_db_hdr,df_db_evl,ol):
    data_dic = {'proyecto':ol,
                'source': 'db',
                 'invitaciones_seg': 0,
                 'provedores_ruc_seg': 0,
                 'proveedores_name_seg': 0,
                 'iniciaron_proceso_seg': 0,
                 'no_participa_seg': 0,
                 'otros_estados_seg': 0,    #-----------------------
                
                 'iniciaron_proceso_hdr': 0,
                 'entregados_hdr': 0,
                 'otros_estados_hdr': 0,
                'linea_adicional': 0,
                 'lineas_principal': 0,
                 'evaluaciones_elv': 0,
                 'avg_puntaje_elv': 0,
                 'informes_inf': 0
               }
    data_dic['invitaciones_seg'] = df_db_seg.shape[0]
    data_dic['provedores_ruc_seg'] = data_db_seg['ciades'].count()
    data_dic['proveedores_name_seg'] = data_db_seg['ciaidenuminv'].count()
    
    data_seg =  data_db_seg.groupby(data_db_seg['maedetcod'].to_list()).size()
    try:
        data_dic['iniciaron_proceso_seg'] = data_seg[8]
    except IndexError:
        data_dic['iniciaron_proceso_seg'] = 0
        
    try:
        data_dic['no_participa_seg'] = data_seg[3]
    except IndexError:
        data_dic['no_participa_seg'] = 0
    
    data_dic['otros_estados_seg'] = data_dic['invitaciones_seg'] - data_dic['iniciaron_proceso_seg']
    
    data_dic['iniciaron_proceso_hdr'] = data_db_hdr.shape[0]
    data_dic['entregados_hdr'] = data_db_hdr[data_db_hdr['maecabestcod'] == '002-HOMEST-5']['ciaidenum_mandante'].count()
    data_dic['otros_estados_hdr'] =  data_dic['iniciaron_proceso_hdr'] - data_dic['entregados_hdr']
    data_dic['linea_adicional'] = data_db_hdr[data_db_hdr['maecabtipcod'] == '002-HOMTIP-7']['ciaidenum_mandante'].count()
    data_dic['lineas_principal'] = data_dic['entregados_hdr'] - data_dic['linea_adicional']
    
    try:
        data_dic['evaluaciones_elv'] = data_db_evl['num_eval'].to_list()[0]
    except:
        data_dic['evaluaciones_elv'] = 0
        
    data_dic['avg_puntaje_elv'] = int(data_db_evl['puntaje'].to_list()[0])
    
    data_dic['informes_inf'] = 0
    
    
    return data_dic

get_db_info(data_db_seg,data_db_hdr,data_db_evl,'245169')
```
